```python
data = pd.read_csv("https://cdn.buenosaires.gob.ar/datosabiertos/datasets/arbolado-publico-lineal/arbolado-publico-lineal-2017-2018.csv")
data.head()
```

```python
plt.scatter(data['long'],data['lat'])
```

```python
arbolado = data
```

```python
poco_arbolado = arbolado.sample(10)
```

```python
poco_arbolado
```

```python
sns.scatterplot(data = poco_arbolado, x='long', y='lat', hue='nombre_cientifico')
```

```python
nombres_cientificos = arbolado['nombre_cientifico'].unique()
```

```python
nombres_cientificos.sort()
```

```python
especies_seleccionadas = nombres_cientificos[(nombres_cientificos>"Citr") & (nombres_cientificos<"Co")] 
```

```python
especies_seleccionadas
```

```python
mascara = arbolado['nombre_cientifico'].isin(especies_seleccionadas)
```

```python
citricos = arbolado[mascara]
```

```python
citricos
```

```python
sns.scatterplot(data=citricos, x='long', y='lat', hue='nombre_cientifico')
```

```python
citricos['comuna'].value_counts()
```

```python
sns.countplot(data=citricos, x='comuna', hue='nombre_cientifico')
```

```python
import plotly.express as px
#df = px.data.gapminder()

fig = px.bar(citricos, x="nombre_cientifico", y='altura_arbol')
fig.show()
```
