```python
# Queremos hacer una programa que me diga si el numero que ingreso mayor o menor a diez
numero_ingresado = input()
numero_ingresado = int(numero_ingresado)

if numero_ingresado > 10:
  print (f"El numero {numero_ingresado} es mayor que 10")
else:
  if numero_ingresado < 10:
    print (f"El numero {numero_ingresado} es menor que 10")
  else:
    print (f"El numero {numero_ingresado} es igual a 10")
print ("Fin!")
```

Ahora el objetivo es un programa que permita introducir nombres y los vaya agregando a una lista. Y que despues elija uno al azar.

```python
nombres = []
nombre_nuevo = ""
while nombre_nuevo != "salir":
  print ("Escriba un nombre")
  nombre_nuevo = input()
  nombres.append(nombre_nuevo)
nombres.pop()
```
