```python
# Un numero es primo solo si es divisible por 1 y por si mismo (y 1 no es si mismo?)
numero = input("Escriba el numero entero que quiere saber si es primo: ")
numero = int(numero)
print(numero)
if numero < 1:
  print ("No tiene sentido preguntar si cero o un numero negativo es primo")
else:
  lista_divisores = []
  divisor_a_testear = 1
  while divisor_a_testear <= numero:
    if (numero // divisor_a_testear) * divisor_a_testear == numero:
      lista_divisores.append(divisor_a_testear)
    divisor_a_testear = divisor_a_testear + 1
  if len(lista_divisores) == 2:
    print ("El numero es primo!")
  else:
    print (f"El numero no es primo, porque tiene todos estos divisores: {lista_divisores}")

```

```python
class Persona:

  def __init__(self, nombre, apellido, edad) -> None:
    self.nombre = nombre
    self.apellido = apellido
    assert type(edad) == int, "La edad debe ser un numero entero"
    self.edad = edad

  def saludar(self):
    print (f"Hola, como andas, yo soy {self.nombre} {self.apellido}")

  def saludar_a_otra_persona(self, persona):
    assert isinstance(persona,Persona), "Solo se puede saludar a un objeto de tipo Persona"
    print (f"Hola {persona.nombre} {persona.apellido}, yo soy {self.nombre} {self.apellido}")

  def __gt__(self, other):
        return(self.edad > other.edad)
```

```python
pepe = Persona("Sapo", "Pepe", 10)
```

```python
persona2 = Persona("Antonio", "Diaz", 8)
```

```python
pepe.saludar()
```

```python
persona2.saludar()
```

```python
pepe.saludar_a_otra_persona(persona2)
```
