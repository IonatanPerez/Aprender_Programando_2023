```python
def calculadora(num1, num2, operation):
    if operation == "suma":
        return num1 + num2

    elif operation == "resta":
        return num1 - num2

    elif operation == "multiplicación":
        return num1 * num2

    elif operation == "división" and num2 != 0:
        return num1 / num2

    else:
        print("Operación no valida")

    return

calculadora(1, 1, "suma")
calculadora(4, 2, "división")
```
