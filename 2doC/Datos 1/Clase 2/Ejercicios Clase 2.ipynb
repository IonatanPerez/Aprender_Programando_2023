#### Ejercicios para practicar

Si queres seguir practicando las estructuras de control que vimos en la clase, Condicionales, Iteraciones y Funciones, animate a hacer estos ejercicios extra optativos:

1. Como en una playa, creá una estructura condicional donde si la bandera es igual a "roja", imprima "prohibición de baño", si es "roja con negro", imprima "mar peligroso"; si es "amarilla con negro", imprima "mar dudoso" y si es "celeste" imprima "mar calmo"

2. Creá una estructura condicional que indique que si la bandera es "roja" **o** "llueve", imprima "no puedo entrar al mar"

3. Creá una estructura condicional que indique que si "no llueve" **y** la bandera es "celeste" imprima "puedo entrar al mar"

4. Creá una lista de ciudades. Crear una estructura de iteración donde cuente la cantidad de ciudades en la lista

5. Creá una lista de números y crear una estructura de iteración donde se vayan multiplicando e imprima el resultado final

6. Creá una función que reciba como argumento 2 números y los multiplique

7. Creá una función que reciba el color de una bandera e imprima el significado (combinar función con estructura condicional)
