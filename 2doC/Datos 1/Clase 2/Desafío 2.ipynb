### Desafío 2 - Estructuras de Control

Vamos a poner en práctica lo que aprendimos de estructuras de control creando estructuras condicionales, iteraciones y funciones:

1. Definí 2 variables con números (num1 y num2). Crear una estructura condicional donde si los dos números son iguales se imprima "los dos números son iguales" (*if... else...*)

2. Creá una estructura condicional donde si "no llueve", "podemos ir a la plaza" y sino "debemos quedarnos en casa" (*if... else...*)

3. Definí una lista con 3 números. Crear una estructura iterativa donde se imprima cada número +2 (*for... in...*)

4. Creá una función que dado 2 números (parámetros), imprima el resultado de dividir el primer nùmero por el segundo (*def*)

5. Creá una función que reciba 2 números e incluya un condicional donde si el segundo número es 0 imprima "no se puede realizar esta operación" y sino realice una división entre el primer y el segundo número.
