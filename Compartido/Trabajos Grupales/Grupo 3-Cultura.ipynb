# Explorando las columnas del dataset


1.  id	= número de identificación
2. title = nombre del libro
3. link = link a la página en goodreads del libro
4. series = ¿pertenece a una serie?
5. cover_link = link a la portada del libro
6. author = autor/a
7. author_link = link al perfil del autor/autora
8. rating_count = cantidad de usuarios que puntuaron el libro
9. review_count = cantidad de usuarios que reseñaron el libro
10. average_rating = puntuación promedio del libro
11. five_star_ratings = cantidad de 5 estrellas que recibió
12. four_star_ratings = cantidad de 4 estrellas que recibió
13. three_star_ratings = cantidad de 3 estrellas que recibió
14. two_star_ratings = cantidad de 2 estrellas que recibió
15. one_star_ratings = cantidad de 1 estrellas que recibió
16. number_of_pages = cantidad de páginas
17. date_published = fecha de publicación
18. publisher	= editorial que publicó el libro
19. original_title = título original
20. genre_and_votes = usuarios que votaron por el libro para cierto género, y cantidad de votos
21. isbn = identificador editorial
22. isbn13 = identificador editoria
23. asin = identificador editoria
24. settings = lugar donde transcurre
25. characters = personajes
26. awards = premios
27. amazon_redirect_link = link a Amazon
28. worldcat_redirect_link
29. recommended_books
30. books_in_series
31. description = descripción del libro
