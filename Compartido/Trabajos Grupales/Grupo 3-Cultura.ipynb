##Podemos ver que hay 846 libros con un puntaje de rating de 5 puntos, por lo que no hay uno solo con el mejor puntaje.

```python
ordenados=data.loc[mejor_libro]
sns.histplot(data=ordenados, y='five_star_ratings')
libro_con_ratings=ordenados["five_star_ratings"].max()
libro_con_ratings=int(libro_con_ratings)
libro_con_mas_5_estrellas=ordenados.loc[(ordenados["five_star_ratings"]==libro_con_ratings),"title"]

print(f"El libro con más votos de 5 estrellas es{libro_con_mas_5_estrellas}")
```
