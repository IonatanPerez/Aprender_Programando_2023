```python
frutas = ["Manzana", "Banana", "Frutilla","Durazno"]

fruta = int(input("Fruta: "))


if fruta < len(frutas):
  print(f"La fruta {fruta} es {frutas[fruta-1]}")
else:
  print(frutas[fruta-1])


```

```python
frutas = [] # Lista de frutas

fruta_nueva = input("Fruta nueva: ")

while fruta_nueva != "salir":
  frutas.append(input(fruta_nueva))

frutas.pop("salir")
```
