```python
materias = ["Matematica", "Fisica", "Lengua"] # Lista de materias

seleccionar_materia = int(input("Numero de materia: ")) # Numero de materia a seleccionar
print(f"Materia: {materias[seleccionar_materia - 1]}") # Imprime la materia

materia_nueva = input("Materia nueva: ") # Le doy una materia
materias.append(materia_nueva) # AÃ±ade la materia a la lista

print(materias) # Imprime la lista de materias

materias.pop(0) # Elimino la primer materia

print(materias) # Imprime las materias nuevamente
```
