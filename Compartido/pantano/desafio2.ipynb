```python
# 2.
print("Queremos ir a la plaza, sin embargo debemos saber si lloverà o no.")
sionocompae = int(input("Si va a llover, entonces introduce el numero 1, si no va a llover, introduce el numero 0"))
if sionocompae == 0:
  print("¡Yey!, entonces podremos ir.")
elif sionocompae == 1:
  print("Oh, en ese caso no podremos ir.")
```

```python
# 3.
numbers = []
new_number = ""
print ("Introduce 3 números.")
while len(numbers) < 3:
  print("Escribe un número:")
  new_number = int(input())
  numbers.append(new_number)
for elemento in numbers:
  print(elemento + 2)
```

```python
# 3. A shorter and easier way to do the same, without for.
numbers = []
new_number = ""
print ("Introduce 3 números.")
while len(numbers) < 3:
  print("Escribe un número:")
  new_number = int(input())
  numbers.append(new_number + 2)
print(numbers)
```

```python
# 4.
print("Introduzca 2 números.")
number1 = int(input("Primer número:"))
number2 = int(input("Segundo número:"))
number3 = number1 // number2
print("El resultado de dividir " + str(number1) + " entre " + str(number2) + " es " + str(number3))
```
