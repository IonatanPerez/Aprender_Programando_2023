1. Hacer un filtro para seleccionar solo los murales realizados en el año 1939

2. Hacer un filtro para seleccionar solo los murales que Luis Seoane. ¿Cuántos murales realizó? ¿Cuál fue el primero?¿Y el úlitmo?

3. Hacer un filtro para seleccionar los murales que NO son realizados con la Técnica "MURAL CERAMICO"

```python
import pandas as pd
murales = pd.read_csv("https://cdn.buenosaires.gob.ar/datosabiertos/datasets/ministerio-de-cultura/murales/murales.csv", sep=";") # Srry, I prefer using web
```

```python
# 1
murales[ murales["F_E"] == 1939 ]
```

```python
# 2
a = murales[ murales["AUTORES"] == "Luis Seoane".upper() ] # filtro
a.shape[0] # cuantos murales realizo

b = a[ a["F_E"] != 0 ] # cual fue el primero
b [ b["F_E"] == b["F_E"].min() ]

a[ a["F_E"] == a["F_E"].max() ] # y el ultimo
```

```python
# 3
murales[ murales["TECNICA"] != "MURAL CERAMICO" ]
```

```python
# 4
murales[ murales["TECNICA"] == "OLEO SOBRE SOPORTE RIGIDO" ]
```

```python
# 5
author = list(dict(murales["AUTORES"].value_counts()))[0]
murales[ murales["AUTORES"] == author ]
```
