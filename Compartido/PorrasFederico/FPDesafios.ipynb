1. Hacer un filtro para seleccionar solo los murales realizados en el año 1939

2. Hacer un filtro para seleccionar solo los murales que Luis Seoane. ¿Cuántos murales realizó? ¿Cuál fue el primero?¿Y el úlitmo?

3. Hacer un filtro para seleccionar los murales que NO son realizados con la Técnica "MURAL CERAMICO" 

4. Generar un dataset filtrando los Murales realizados con la Técnica "OLEO SOBRE SOPORTE RIGIDO"

5. Filtrar el DataFrame seleccionando los Murales realizados por el artista que más Murales realizó

```python
import pandas as pd
murales = pd.read_csv("https://cdn.buenosaires.gob.ar/datosabiertos/datasets/ministerio-de-cultura/murales/murales.csv", sep=";") # Srry, I prefer using web
```

```python
# 1
murales[ murales["F_E"] == 1939 ]
```

```python
# 2
a = murales[ murales["AUTORES"] == "Luis Seoane".upper() ] # filtro
a.shape[0] # cuantos murales realizo

b = a[ a["F_E"] != 0 ] # cual fue el primero
b [ b["F_E"] == b["F_E"].min() ]

a[ a["F_E"] == a["F_E"].max() ] # y el ultimo
```

```python
# 3
murales[ murales["TECNICA"] != "MURAL CERAMICO" ]
```

```python
# 4
murales[ murales["TECNICA"] == "OLEO SOBRE SOPORTE RIGIDO" ]
```

```python
# 5
author = list(dict(murales["AUTORES"].value_counts()))[0]
murales[ murales["AUTORES"] == author ]
```

### Desafio 5: Visualizaciones

Para este Desafío vamos a utilizar el Dataset del Arbolado que venimos utilizando


1. Hacer un gráfico que muestre cuántos arboles de *Tipuana tipu*hay en cada comuna. Primero realizar un filtro por *nombre_científico* y luego realizar el gráfico.

2. Hacer un gráfico que muestre el ancho de los árboles por comuna ¿Qué comunas tienen arboles más anchos?

3. Realizar un gráfico que muestre la distribución del ancho de los árboles, ¿Qué ancho tienen la mayoría de los arboles?

4. ¿Cuál es la calle con más arboles? Realizar un *value_counts()*. Realizar un gráfico *displot* con la *altura_arbol* y el *diametro_altura_pecho* de los arboles de esa calle (para eso se debe filtrar el dataset)

```python
import pandas as pd
data = pd.read_csv("https://cdn.buenosaires.gob.ar/datosabiertos/datasets/arbolado-publico-lineal/arbolado-publico-lineal-2017-2018.csv")
```

```python
import seaborn as sns
```

```python
data.head()
```

```python
firstPointFilter = data[ data["nombre_cientifico"] == "Tipuana tipu" ]
```

```python
# 1
sns.histplot(data=firstPointFilter, x="comuna")
```

```python
# 2
sns.scatterplot(data=data, x="comuna", y="diametro_altura_pecho", alpha=0.01)
```

```python
# 3
sns.histplot(data=data, x="diametro_altura_pecho")
```

```python
# 4 ¿Cuál es la calle con más arboles? Realizar un value_counts(). Realizar un gráfico displot con la altura_arbol y el diametro_altura_pecho de los arboles de esa calle (para eso se debe filtrar el dataset)
street = data["calle_nombre"].value_counts().index[0]
sns.displot(data=data[ data["calle_nombre"] == street ], y="altura_arbol", x="diametro_altura_pecho")
```
