````python
myString = "Esto es un texto"

def act1():
  name = input("Name: ")
  print("Last letter:", name[-1])

def act2():
  listNames = ["Isabella", "Daniela", "Luciano", "Joaquin", "Milena"]
  num = input("Choose the Nº of name that you want to be printed: ")
  try:
    num = int(num)
    print(listNames[num-1])
  except ValueError:
    print("Bad entry, the Nº have to be between 1 and 5 (there isn't more names)"); return "ValueError"

def act3():
  num = input("Num: ")

  try: num = int(num)
  except ValueError: print("Bad entry, the Nº have to be an integer"); return "ValueError"

  if num > 10: print(num, "is greater than 10")
  elif num == 10: print(num, "is equal to 10")
  else: print(num, "is lower than 10")

  print("End.")

def act4(showNames=False):
  names = []
  while True:
    print('"exit" to exit the program | List:', names)
    newName = input("Nuevo nombre: ")
    if newName == "exit": break
    print("Hi", newName)
    names.append(newName)

  if showNames:
    for i in names: print("Hi", i)

  return names

act4(showNames=True)
````

### **Notas de la clase**
* `datos.dropna()` elimina todas las filas con Nans
* `datos.dropna(axis=1)` elimina todas las columnas con Nans
* `datos["edades"].isna().sum()`
* `edades_no_nan = datos["edades"][~datos["edades"].isna()]`
 * el ~ sirve para negar cada uno de los elementos (el ! negaría la lista completa, ¿y que es negar una lista?). '!' Niega el objeto, '~' (alt+126) se mete adentro del objeto.
 * lo mismo pasa con 'or' y '|', y 'and' y '&'
* `datos.loc[~mascara_no_nulos, "edades"] = datos["edades"].mean()` Rellenar las personas que tienen edad NaN con la edad promedio
* `data.replace("regular ", "Regular", inplace=True)`

````python
arboladoData["ubicacion_plantera"].replace("regular", "Regular", inplace=True)
arboladoData["ubicacion_plantera"].replace(" regular", "Regular", inplace=True)
arboladoData["ubicacion_plantera"].replace("Regular ", "Regular", inplace=True)

arboladoData["ubicacion_plantera"].replace("Och ", "Ochava", inplace=True)
arboladoData["ubicacion_plantera"].replace("Ochva", "Ochava", inplace=True)
arboladoData["ubicacion_plantera"].replace("ochava", "Ochava", inplace=True)

print(arboladoData["ubicacion_plantera"].unique())
arboladoData["ubicacion_plantera"].value_counts()
````

````python
import numpy
````

````python
nanList = (
    "O", "Och", 'o', 'Ochava/Fuera Línea', 'Fuera de Línea, Ochava'
)

repeatedList = (
       'Fuera Línea,Ochava', 'Cantero Grande', 'Ocupada',  'Elevada',
        'Subocupada', 'ochava', 'Fuera Nivel', 'Sobreocupada',
       'Fuera Línea/Ochava'
)

toNanDict = {i:numpy.nan for i in nanList}
toReplaceDict = {i:"Otro" for i in repeatedList}

# Making a bigger dict with the info off all
reemplazo = {}
reemplazo.update(toNanDict)
reemplazo.update(toReplaceDict)

arboladoData["ubicacion_plantera"].replace(reemplazo, inplace=True)

print(arboladoData["ubicacion_plantera"].unique())
arboladoData["ubicacion_plantera"].value_counts()
````
