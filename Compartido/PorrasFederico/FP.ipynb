```python
myString = "Esto es un texto"

def act1():
  name = input("Name: ")
  print("Last letter:", name[-1])

def act2():
  listNames = ["Isabella", "Daniela", "Luciano", "Joaquin", "Milena"]
  num = input("Choose the Nº of name that you want to be printed: ")
  try: 
    num = int(num)
    print(listNames[num-1])
  except ValueError: 
    print("Bad entry, the Nº have to be between 1 and 5 (there isn't more names)"); return "ValueError"

def act3():
  num = input("Num: ")
  
  try: num = int(num)
  except ValueError: print("Bad entry, the Nº have to be an integer"); return "ValueError"
  
  if num > 10: print(num, "is greater than 10") 
  elif num == 10: print(num, "is equal to 10")
  else: print(num, "is lower than 10")

  print("End.")

def act4(showNames=False):
  names = []
  while True:
    print('"exit" to exit the program | List:', names)
    newName = input("Nuevo nombre: ")
    if newName == "exit": break
    print("Hi", newName)
    names.append(newName)

  if showNames: 
    for i in names: print("Hi", i)

  return names

act4(showNames=True)
```

#### Busqueda de datos:
* Medibles → En algún lado tiene que haber registro 
  * *Ej: Cuando se pasa la sube por el molinete, queda registro en algún lado. ¿Que lado? Registro del gobierno / Empresa de transporte*
* Posibilidad de acceso → ¿El poseedor de la info la comparte?¿Porque? *(muchas veces no)*
  * Datos Abiertos → Los gobiernos tienen información que deben dar por ley
  * *Ej: Un ministro de transporte necesita saber cuantas personas viajan, no quienes son*
* Objetivo de recabarlos → ¿Que necesito saber de las personas? *quienes, cuantos, en que rango de edad, etc.*
  * Circunstancias → Hay casos en los que no importa si son datos personales, se pueden recabar igual (ej: caso judicial).
