Desafío 1 - Listas

Hacer una lista que contenga 3 materias de la escuela, 
Eliminar el primer registro de la lista, 
Agregar una materia más al final, 
Seleccionar el segundo elemento de la lista.

```python
materias = ["historia" , "matematica" , "quimica"]
print (f"Las lista inicial de tres materias es: {materias}")
print (f"La segunda materia de mi lista es: {materias [1]}")
print ("Que materia quieres agregar a la lista? ")
nueva_materia = input ()
materias.append (nueva_materia) #agrego la nueva materia
print (f"La nueva lista de materias es: {materias}") #imprimo la lista de materias con la nueva incluida
materias.pop (0) #elimina el ultimo de la lista, si pongo cero elimina el 1ero
print (f"La nueva lista sin el primer registro es: {materias}")
```

```python
print (materias [1]) #seleccionar el 2do elemento de la lista
```
