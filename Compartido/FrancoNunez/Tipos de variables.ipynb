```python
nombres = ["Isabella", "Daniela", "Luciano", "Joaquin", "Milena"]
print ("Elija el numero de persona que quiere seleccionar")
seleccion = input()
seleccion = int(seleccion)
seleccion = seleccion - 1
if seleccion < len(nombres):
  print (f"El nombre seleccionado es {nombres[seleccion]}")
else:
  print ("No hay tantos nombres en la lista.")
```

```python
# Queremos hacer un programa que me diga si el numero que ingreso es mayor o menor a diez
numero_ingresado = input()
numero_ingresado = int(numero_ingresado)

if numero_ingresado > 10:
  print (f"El numero {numero_ingresado} es mayor que 10")
else:
  if numero_ingresado < 10:
    print (f"El numero {numero_ingresado} es menor que 10")
  else:
    print (f"El numero {numero_ingresado} es igual que 10")
print ("Fin!")
```

Ahora el objetivo es un programa que permita introducir nombres y los vaya agregando en una lista. Y que despues elija uno al azar.

```python
nombres = []
nombre_nuevo = ""
while nombre_nuevo != "salir":
  print ("Escribe un nombre")
  nombre_nuevo = input()
  nombres.append(nombre_nuevo)
nombres.pop()
```
